```python
LEVEL 2 : TASK 1
 # TASK : Restaurant Ratings
# Analyze the distribution of aggregate ratings and determine the most common rating range.
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot a histogram of the aggregate ratings
plt.figure(figsize=(10, 6))
plt.hist(df['Aggregate rating'], bins=10, color='blue', edgecolor='black')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.title('Distribution of Aggregate Ratings')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Determine the most common rating range
rating_counts = pd.cut(df['Aggregate rating'], bins=10).value_counts().sort_values(ascending=False)
most_common_rating_range = rating_counts.idxmax()

most_common_rating_range
```

```python
LEVEL 2 : TASK 2 
# TASK : Cuisine Combination
```

```python
# Identify the most common combinations of cuisines in the dataset.
```

```python
# Count the occurrences of each unique combination of cuisines
cuisine_combinations_counts = df['Cuisines'].value_counts()

# Display the most common combinations of cuisines
most_common_cuisine_combinations = cuisine_combinations_counts.head(10)

most_common_cuisine_combinations
```

```python
 # Determine if certain cuisine combinations tend to have higher ratings.
```

```python
# Group the dataset by the 'Cuisines' column and calculate the average rating for each combination
average_ratings_by_cuisine_combination = df.groupby('Cuisines')['Aggregate rating'].mean()

# Sort the cuisine combinations by their average ratings in descending order
sorted_average_ratings = average_ratings_by_cuisine_combination.sort_values(ascending=False)

# Display the top 10 cuisine combinations with the highest average ratings
top_cuisine_combinations_with_highest_ratings = sorted_average_ratings.head(10)

top_cuisine_combinations_with_highest_ratings
```

```python
import folium
from folium.plugins import MarkerCluster
```

```python
longitude = df['Longitude']
latitude = df['Latitude']
```

```python
# Create a base map centered around an average coordinate (can be set to a specific city or area)
map_center = [latitude.mean(), longitude.mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=12)

# Add each restaurant's location to the map
for lat, lon in zip(latitude, longitude):
    folium.Marker(location=[lat, lon]).add_to(restaurant_map)

# Save the map to an HTML file
restaurant_map.save('restaurants_map.html')

# Display the map in Jupyter Notebook or compatible environment
restaurant_map
```

```python
# Identify any patterns or clusters of restaurants in specific areas.
```

```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Extract latitude and longitude coordinates
coords = df[['Latitude', 'Longitude']].to_numpy()

# Perform DBSCAN clustering
db = DBSCAN(eps=0.01, min_samples=10).fit(coords)
labels = db.labels_

# Add cluster labels to the dataset
df['Cluster'] = labels
```

```python
# Plot clusters
plt.figure(figsize=(10, 6))
plt.scatter(df['Longitude'], df['Latitude'], c=df['Cluster'], cmap='viridis', marker='o')
plt.title('Restaurant Clusters')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
LEVEL 2 : TASK 4 
# TASK : Restaurant Chains 
```

```python
# Identify if there are any restaurant chains present in the dataset.
```

```python
# Count the occurrences of each restaurant name
restaurant_counts = df['Restaurant Name'].value_counts()

# Identify restaurant chains (names that appear more than once)
restaurant_chains = restaurant_counts[restaurant_counts > 1]

restaurant_chains
```

```python
# Analyze the ratings and popularity of different restaurant chains.
```

```python
# Identify restaurant chains by counting occurrences of each restaurant name
restaurant_counts = df['Restaurant Name'].value_counts()
restaurant_chains = restaurant_counts[restaurant_counts > 1].index

# Filter the dataset to include only restaurant chains
chain_restaurants = df[df['Restaurant Name'].isin(restaurant_chains)]

# Group by restaurant name and calculate the average rating and total number of votes for each chain
chain_stats = chain_restaurants.groupby('Restaurant Name').agg({
    'Aggregate rating': 'mean',
    'Votes': 'sum'
}).sort_values(by=['Aggregate rating', 'Votes'], ascending=[False, False])

# Display the top 10 restaurant chains by average rating and popularity
top_chains_by_rating = chain_stats.sort_values(by='Aggregate rating', ascending=False).head(10)
top_chains_by_votes = chain_stats.sort_values(by='Votes', ascending=False).head(10)

top_chains_by_rating, top_chains_by_votes
```

```python
# Convert the 'Votes' column to numeric type if it is not already
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')

# Find the restaurant with the highest number of votes
restaurant_highest_votes = df.loc[df['Votes'].idxmax()]

# Find the restaurant with the lowest number of votes
restaurant_lowest_votes = df.loc[df['Votes'].idxmin()]

restaurant_highest_votes, restaurant_lowest_votes
```
